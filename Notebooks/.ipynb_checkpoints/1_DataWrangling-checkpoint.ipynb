This notebook is for wrangling the bike share data and generating clean data for performing some exploratory data analysis in the future. **Note:** the code below includes machine specific file paths to store the clean data. Please change these file paths appropriately before running the code. 

```python
# Import Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

## Daily Data

```python
# Load the daily csv data
file_path = "/Users/aditigala/Documents/Pacific/Software_Methods/Bikeshare/Data/daily.csv"
daily = pd.read_csv(file_path)
print("Rows, Columns = ", daily.shape)

# Check for NaNs
print("\nThe no. of Nas by column:")
print(daily.isna().sum())

# Plot of bike rentals over time
plt.figure(figsize = (10,6))
date_objects = [pd.datetime.strptime(date, '%Y-%m-%d').date() for date in daily["dteday"]]
plt.plot(date_objects, daily["cnt"])
plt.xticks(rotation = 45)
plt.xlabel('Date')
plt.ylabel('Count')
plt.title("Count of Bike Rentals by Day")
plt.show()
```

```python
# Extract Year, Month, Day from Date column and use as replacement
daily["Year"] = pd.DatetimeIndex(daily["dteday"]).year
daily["Month"] = pd.DatetimeIndex(daily["dteday"]).month
daily["Day"] = pd.DatetimeIndex(daily["dteday"]).day

# Print the first five rows of the data frame
daily.head()
```

```python
# Save the new dataframe as "daily_clean.csv" in /Data folder
daily.to_csv("/Users/aditigala/Documents/Pacific/Software_Methods/Bikeshare/Data/daily_clean.csv")
```

## Hourly Data
